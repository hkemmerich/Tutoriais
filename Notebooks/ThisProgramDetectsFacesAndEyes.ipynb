```
#Description: This program detects faces and eyes
```

```
#Import the open CV Libary
import cv2
```

```
#The Haar classifiers
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier('haarcascade_eye.xml')
```

```
#Read in the img
img = cv2.imread('dwayne_and_james.jpg')
## faces = face_cascade.detectMultiScale(img, 1.3, 5) os dois ultimos paramentos s√£o opcionais
faces = face_cascade.detectMultiScale(img)
```

```
#Print the number of faces found
print('Faces found: ', len(faces))
print('The image height, width and channel: ', img.shape)
print('The coordinates of each face detected: ', faces)
```

```
#Loop onver all the coordinates faces and draw rectangules
for (x, y, w, h) in faces:
   cv2.rectangle(img, (x,y), (x+w, y+h), (0,255,0))
   roi_face = img[y:y+h, x:x+w]
   eyes = eye_cascade.detectMultiScale(roi_face)
   for (ex, ey, ew, eh) in eyes:
     cv2.rectangle(roi_face, (ex,ey), (ex+ew, ey+eh), (255, 0, 0), 2)
```

```
font = cv2.FONT_HERSHEY_SIMPLEX
text = cv2.putText(img, 'Face Detected', (0, img.shape[0]), font, 2, (255,255,255),2)
```
